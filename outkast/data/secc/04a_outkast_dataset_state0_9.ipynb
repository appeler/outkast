```python
import pandas as pd
```

```python
df = pd.read_csv('/opt/data/secc/secc_state_name_age_sex_social_cat_state0_9.csv.gz')
df
```

```python
df.dropna(subset=['last_name'], inplace=True)
df.drop(df[~df.last_name.str.contains('^[a-z]{2,}')].index, inplace=True)
df.drop(df[df.last_name.str.contains('\.')].index, inplace=True)
df
```

* i think it makes sense to only release data where count for last_name/first_name is > 100
* it will take out a bunch of weird names for now

```python
adf = df[df['last_name'].groupby(df['last_name']).transform('size')>100]
adf
```

```python
adf['last_name'].groupby(adf['last_name']).size()
```

```python
gdf = adf.groupby(['state', 'birth_year', 'last_name']).agg({'gender': 'value_counts', 'social_cat': 'value_counts'}).unstack()
```

```python
gdf
```

```python
gdf.reset_index(inplace=True)
gdf
```

```python
gdf.columns = ['_'.join(col).strip() for col in gdf.columns.values]
gdf
```

```python
gdf = gdf[['level_0_', 'level_1_', 'level_2_', 'gender_m', 'gender_f', 'social_cat_c', 'social_cat_o', 'social_cat_t']]
gdf.columns = ['state', 'birth_year', 'last_name', 'n_male', 'n_female', 'n_sc', 'n_other', 'n_st']
gdf
```

```python
gdf.fillna(0, inplace=True)
gdf.astype({'n_male': int, 'n_female': int, 'n_sc': int, 'n_st': int, 'n_other': int})
```

```python
gdf.to_csv('/opt/data/secc/secc_state0_9_year_ln_outkast.csv.gz', index=False, compression='gzip')
```

```python
bdf = gdf.groupby(['last_name']).agg({'n_sc': 'sum', 'n_other': 'sum', 'n_st': 'sum', 'n_male': 'sum', 'n_female': 'sum'}).reset_index()
bdf
```
