```python
import pandas as pd
```

```python
df0_9 = pd.read_csv('/opt/data/secc/secc_state0_9_year_ln_outkast.csv.gz', dtype={'n_male': int, 'n_female': int, 'n_sc': int, 'n_st': int, 'n_other': int})
df10_18 = pd.read_csv('/opt/data/secc/secc_state10_18_year_ln_outkast.csv.gz', dtype={'n_male': int, 'n_female': int, 'n_sc': int, 'n_st': int, 'n_other': int})
df = pd.concat([df0_9, df10_18])
df
```

```python
gdf = df.groupby(['last_name']).agg({'n_sc': 'sum', 'n_other': 'sum', 'n_st': 'sum', 'n_male': 'sum', 'n_female': 'sum'}).reset_index()
gdf
```

```python
sdf = pd.read_csv('secc_state.csv')
sdf
```

```python
odf = df.rename(columns={'state': 'code'}).merge(sdf)
del odf['code']
odf
```

```python
cols = ['state', 'birth_year', 'last_name', 'n_female', 'n_male', 'n_sc', 'n_st', 'n_other']
odf[cols].to_csv('/opt/data/secc/secc_all_state_year_ln_outkast.csv.gz', index=False, compression='gzip')
```
